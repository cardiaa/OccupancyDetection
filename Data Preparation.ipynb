```python
import pandas as pd
import csv
def txt_to_csv(filename):
    with open(filename + ".txt", 'r') as in_file:
        stripped = (line.strip() for line in in_file)
        lines = (line.split(",") for line in stripped if line)
        with open(filename + '.csv', 'w') as out_file:
            writer = csv.writer(out_file)
            writer.writerows(lines)
        

def change_date(df):
    new_col = []
    for index, data in enumerate(df['date']):
        _, data, _ = df['date'][index].split('"')
        new_col.append(data)
    df['date'] = new_col
```

```python
txt_to_csv('datatraining')
txt_to_csv('datatest')
txt_to_csv('datatest2')
df = pd.read_csv("datatraining.csv")
df1 = pd.read_csv('datatest.csv')
df2 = pd.read_csv('datatest2.csv')
```

## "Date" tranformation

Creation of new features from 'date'

```python
df['date']
```

```python
date=[]
time=[]
for i in df['date']:
    d, t = i.split(" ")
    #we are interested only in day information
    _,d = d.split("2015-02-")
    date.append(int(d))
    time.append(t)
```

```python
set(date)
```

```python
# time separation
hours = []
minutes = []
for t in time:
    h, m, s = t.split(":")
    
    if int(s) == 59:
        if int(m) == 59:
            #print(minutes[len(minutes)])
            minutes.append(0)
            h = int(h)
            h +=1
        else:
            minutes.append(int(m) + 1)
            
    else:
        minutes.append(int(m))
    hours.append(int(h))
```

```python
new_time = []
for i in range(len(hours)):
    #print("{}:{}".format(hours[i], minutes[i]))
    new_time.append((hours[i]*60)+minutes[i])
```

```python
#14*60 + 19 = 859 (numero di minuti a cui iniziano a partire le registrazioni. questo perche' il primo dato lo abbiamo alle 14:19 del lunedi' 02-02-15)
new_time[0]
```

```python
for i in range(len(date)):
    print("date: {}, time: {:2}:{:<2}, time in min: {:4}"
          .format(date[i], hours[i], minutes[i], new_time[i]))
```

```python
# aggiungiamo le variabili nuove al dataset
df["day"] = date
df['hour'] = hours
df['minute'] = minutes
df['day_minute'] = new_time
```

```python
cumulative_hour =[]
cumulative_minute = []

starting_hour = df['day'][0]*24 + df['hour'][0]
starting_minute = df['minute'][0]

for index, row in df.iterrows():
    cumulative_hour.append(((df['day'][index]*24) + df['hour'][index]) - starting_hour) # last element of the formula since observation started from day 4 hour 17
    cumulative_minute.append((((df['day'][index]*24) + df['hour'][index]) - starting_hour)*60 + (df['minute'][index]) - starting_minute) # 1071 is the value of day_minute at the first observation

df['cumulative_hour'] = cumulative_hour
df['cumulative_minute'] = cumulative_minute
```

```python
df['weekend'] = 0
for index, row in df.iterrows():
    if row['day']%7 == 0 or row['day']%7 == 1:
        df['weekend'][index] = 1
```
